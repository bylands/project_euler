```python
long = [1, 3, 5, 7, 8, 10, 12]
short = [2, 4, 6, 9, 11]

def days_in_year_till(year, month, day):
    count = 0

    for m in range(1, month):
        if m in long:
            count += 31
        else:
            count += 30
    if month > 2:
        if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
            count -= 1
        else: 
            count -= 2
    count += day

    return count
            
def days_in_year_from(year, month, day):
    count = 0

    for m in range(month, 12):
        if m+1 in long:
            count += 31
        else:
            count += 30
        if m < 2:
            if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
                count -= 1
            else: 
                count -= 2
    
    if month in long:
        count += 31 - day
    elif month != 2:
        count += 30 - day
    else:
        if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
            count += 29 - day
        else: 
            count += 28 - day
    
    return count

def days_in_year(year):
        if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
            return 366
        else:
            return 365

def days_between(year1, month1, day1, year2, month2, day2):
    if year2 < year1:
        print('second date has to be after first date')
        return -1
    elif year1 == year2:
        if month2 < month1:
            print('second date has to be after first date')
            return -1
        elif month1 == month2:
            if day2 < day1:
                print('second date has to be after first date')
                return -1
            else:
                return day2 - day1
        else:
            return days_in_year_till(year2, month2, day2) - days_in_year_till(year1, month1, day1)
    
    count = 0
    if year2 - year1 > 1:
        for y in range(year1+1, year2):
            count += days_in_year(y)
    count += days_in_year_from(year1, month1, day1) + days_in_year_till(year2, month2, day2)
    return count

def weekday(year, month, day):
    daynames = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

    return daynames[days_between(1900, 1, 1, year, month, day) % 7]
```

```python
solution = 0

for y in range(1901, 2001):
    for m in range(1, 13):
        if weekday(y, m, 1) == 'Sun':
            solution += 1

print(f'During the twentieth century {solution} Sundays fell on the first of the month.')
```

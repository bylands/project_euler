```python
import csv

with open('euler_ressources/0089_roman.txt') as csvfile:
    romans = [row[0] for row in csv.reader(csvfile)]
```

```python
numerals = ['M', 'D', 'C', 'L', 'X', 'V', 'I']
fivers = ['D', 'L', 'V']
tenners = ['M', 'C', 'X', 'I']
order = {'I': 0, 'V': 1, 'X': 2, 'L': 3, 'C': 4, 'D': 5, 'M': 6}
values = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}


def roman_to_decimal(roman: str) -> int:

    length = len(roman)
    value = 0

    for i, r in enumerate(roman):
        if i < length-1:
            if order[roman[i+1]] > order[r]:
                value -= values[r]
            else:
                value += values[r]
        else:
            value += values[r]
        
    return value


def decimal_to_roman(n: int) -> str:
    roman = ''
    for i, r in enumerate(tenners):
        multiple = 0
        while n - (v:=values[r]) >= 0:
            roman += r
            multiple += 1
            n -= v
        if r != 'M' and multiple >= 4:
            if multiple < 5:
                roman = roman[:-multiple] + r + fivers[i-1]
            elif multiple < 9:
                roman = roman[:-multiple] + fivers[i-1] + r*(multiple-5)
            elif multiple == 9:
                roman = roman[:-multiple] + r + tenners[i-1]
        # if len(roman) >= 4 and len(set(roman[-4:])) == 1 and r != 'M':
            # if len(roman) >= 5:
            #     if roman[-5] not in fivers:
            #         roman = roman[:-4] + r + numerals[i-1]
            #     else:
            #         roman = roman[:-5] + r + numerals[i-2]
            # else:
            # roman = roman[:-4] + r + numerals[i-1]

    return roman
```

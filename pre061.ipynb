```python
'''
Project Euler Problem 061: Cyclical Figurate Numbers
'''
```

```python
from itertools import permutations
```

```python
def poly(p, n):
    return ((p-2)*n*n-(p-4)*n)//2


def triangle(n):
    return n*(n+1)//2


def square(n):
    return n*n


def pentagonal(n):
    return n*(3*n-1)//2


def hexagonal(n):
    return n*(2*n-1)


def heptagonal(n):
    return n*(5*n-3)//2


def octagonal(n):
    return n*(3*n-2)


def left(n, split=2):
    return n // 10**split


def right(n, split=2):
    return n % 10**split


def get_matches(n, sets, index):
    return [x for x in sets[index] if right(n) == left(x)]
```
